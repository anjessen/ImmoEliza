### Group by locality into Commune

```python
'''Get the dataset of Communes by postal code here :
https://public.opendatasoft.com/explore/dataset/liste-des-codes-postaux-belges-fr/table/?flg=fr
'''
postal_codes = pd.read_csv('./Datas/liste-des-codes-postaux-belges-fr.csv', sep=';')
postal_codes.head(10)
```

```python
# we only want to keep postal code and principal commune
commune_index = postal_codes[['Code postal', 'Commune principale']]
commune_index.head(10)
```

```python
commune_index = commune_index.copy()
commune_index.loc[1,'Commune principale'] = 'ROCHEFORT'
commune_index.loc[2763, 'Commune principale'] = 'SAINT-JOSSE-TEN-NOODE'
commune_index.loc[715, 'Commune principale'] = 'SAINT-JOSSE-TEN-NOODE'
commune_index.loc[748, 'Commune principale'] = 'SCHAERBEEK'
commune_index.loc[1389, 'Commune principale'] = 'SCHAERBEEK'
commune_index.loc[2744, 'Commune principale'] = 'SCHAERBEEK'
```

```python
commune_index['Commune principale'].isna().sum()
```

```python
'''mask = commune_index['Commune principale'].isna()
commune_index.loc[mask]'''
```

```python
commune_index.head(10)
```

```python
commune_index.dropna(axis=0, inplace=True, subset=['Commune principale'])
```

```python
commune_index['Commune principale'].isna().sum()
```

```python
commune_index[commune_index['Code postal']==8957].index.values
commune_index.loc[2114]
```

```python
commune_index = commune_index.set_index('Code postal')
commune_index.head(5)
```

```python
dict_commune = commune_index.to_dict()
dict_commune
```

```python
# adding the commune feature to the base dataset
df['commune'] = [dict_commune['Commune principale'][pc] for pc in df['locality']]
```

```python
df.commune.isnull().sum()
```

```python
df[['commune','locality']].head(10)
```

```python
# order and rank the communes by price / m² 
commune_rank_dict = df.groupby('commune')['price_per_m²'].mean().sort_values().rank().to_dict()
commune_rank_dict
```

```python
# save the dict to use it in the API 
import csv

dict_to_save = commune_rank_dict
w = csv.writer(open("./Datas/commune_rank.csv", "w"))
for key, val in dict_to_save.items():
    w.writerow([key, val])
```

```python
# add rank_commune to the base dataset 
df['rank_commune'] = [commune_rank_dict[commune_name] for commune_name in df['commune']]
```
